**c) From the above hypothesis, I have compelling reason to believe that old users might not motivated enough to use the product. I wrote a query to group the activity by the age of the users(the difference between sign up and engagement date) using the product. I have plotted the below graph to prove my hypothesis.**

```python
%sql

/* 2nd part of question on weekly users converted to months (graph) */

select 
DATE_TRUNC('week', e.occurred_at) as week,
(CASE
    WHEN FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) >= 0 and FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) < 4 THEN '0 month'
    WHEN FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) >= 4 and FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) < 8 THEN '1 month'
    WHEN FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) >= 8 and FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) < 12 THEN '2 months'
    ELSE '3+ months'
  END) AS history, count(e.event_type) as user_activity 
from events e inner join users u
on e.user_id = u.user_id
where e.event_type = 'engagement'
group by 1, 2 order by 1, 2;
```

**The above graph clearly displays a significant drop in the user activity for users who are using the product for more than 3 months. This indicates that the old users might not be liking the product/application or they found an alternative solution for it.**

```python
%sql

/* 2nd part of question on weekly users converted to months (graph) */

select 
DATE_TRUNC('week', e.occurred_at) as week,
(
  CASE
    WHEN FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) >= 0 and FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) < 4 THEN '0 month'
    WHEN FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) >= 4 and FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) < 8 THEN '1 month'
    WHEN FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) >= 8 and FLOOR(DATEDIFF(e.occurred_at, u.created_at)/7) < 12 THEN '2 months'
    ELSE '3+ months'
  END
) AS history, count(e.event_type) as user_activity 
from events e
inner join users u
on e.user_id = u.user_id
where e.event_type = 'engagement'
group by 1, 2
order by 1, 2;
```

**d) One more assumption I had in mind is that user activity might depend on the day when the user interacts with the product. So I wrote a SQL query to check my hypothesis by grouping the user activity based on weekdays and weekends. I plotted a graph based on the data I got from executing the query.**

```python
%sql

/* 2nd part of question on days of week (graph) */

SELECT
  (CASE
    WHEN dayofweek(occurred_at) = 1  then 'Sunday'
    WHEN dayofweek(occurred_at) = 2  then 'Monday'
    WHEN dayofweek(occurred_at) = 3  then 'Tuesday'
    WHEN dayofweek(occurred_at) = 4  then 'Wednesday'
    WHEN dayofweek(occurred_at) = 5  then 'Thursday'
    WHEN dayofweek(occurred_at) = 6  then 'Friday'
    ELSE 'Saturday' END) AS day, DATE_TRUNC('week', occurred_at) AS date, count(*) AS user_activity
FROM events WHERE event_type = 'engagement' group by 1, 2 order by 1, 2;
```

**The above graph indicates that the user activity is very less on Saturday and Sunday. So I plotted the user activity data to differentiate between weekdays and weekends to reduce the clutter in the pie chart above.**

```python
%sql

/* 2nd part of question on days of week (table) */

SELECT
  (CASE
    WHEN dayofweek(occurred_at) = 1  then 'Sunday'
    WHEN dayofweek(occurred_at) = 2  then 'Monday'
    WHEN dayofweek(occurred_at) = 3  then 'Tuesday'
    WHEN dayofweek(occurred_at) = 4  then 'Wednesday'
    WHEN dayofweek(occurred_at) = 5  then 'Thursday'
    WHEN dayofweek(occurred_at) = 6  then 'Friday'
    ELSE 'Saturday' END) AS day, DATE_TRUNC('week', occurred_at) AS date, count(*) AS user_activity
FROM events WHERE event_type = 'engagement' group by 1, 2 order by 1, 2;
```

```python
%sql

/* 2nd part of question on weekdays and weekends (graph) */

SELECT
  (CASE
    WHEN dayofweek(occurred_at) in (1, 7)  then 'weekends'
    ELSE 'weekdays' END) AS day, DATE_TRUNC('week', occurred_at) AS date, count(*) AS user_activity
FROM events WHERE event_type = 'engagement' group by 1, 2 order by 1, 2;
```

**From the graph shown, we can deduce that the users use the product on weekdays more than weekends. This might be because they use it mostly during their work hours. The drop in the user activity on weekdays started on July 27th. Organizations normally acquire licenses for using third party applications. So one or more organizations using the product could have not renewed the license or opted for some other application. The product might no longer be helpful for the organization growth.**

```python
%sql

/* 2nd part of question on weekdays and weekends (table) */

SELECT
  (CASE
    WHEN dayofweek(occurred_at) in (1, 7)  then 'weekends'
    ELSE 'weekdays' END) AS day, DATE_TRUNC('week', occurred_at) AS date, count(*) AS user_activity
FROM events WHERE event_type = 'engagement' group by 1, 2 order by 1, 2;
```
